##Academic Structure & Student Progress

Data Source: Kaggle - Students' Mental Health Survey

### Research Question
How does academic workload differ across academic years, and how is it related to study satisfaction?

```python
import pandas as pd

df = pd.read_csv("/MentalHealthSurvey.csv")  # ← 네 CSV 경로
df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
df[['academic_year','academic_workload','study_satisfaction']].info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

plt.figure(figsize=(6,4))
sns.boxplot(x='academic_year', y='academic_workload', data=df)
plt.title('Academic Workload by Academic Year')
plt.xlabel('Academic Year')
plt.ylabel('Academic Workload')
plt.show
```

Workload levels differ across academic years, suggesting variation in academic demands as students progress.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='academic_workload', y='study_satisfaction', data=df)
plt.title('Study Satisfaction by Academic Workload')
plt.xlabel('Academic Workload')
plt.ylabel('Study Satisfaction')
plt.show
```

Higher academic workload appears to be associated with lower study satisfaction, though variability within groups.

## Brief Findings
- Academic workload differs across academic years, indicating changes in academic demands as students progress.

- Higher levels of academic workload tend to be associated with lower study satisfaction.

- These patterns suggest that structural academic factors may influence students’ learning experience.

## Limitations

- This analysis is exploratory and based on self-reported survey data.

- Findings do not imply causation and may not generalize beyond the sample.
