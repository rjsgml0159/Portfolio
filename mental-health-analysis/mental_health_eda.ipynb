## Mental Health & Behavioral Patterns

### Introduction & Research Question

This project explores relationships between sleep, stress, and mental health indicators
using a publicly available student mental health survey dataset.

### Dataset Overview
The dataset contains responses from 87 students and includes demographic information,
academic factors, and self-reported sleep and stress-related measures.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("/MentalHealthSurvey.csv")
df.head()
```

```python
df.info()
```

### Exploratory Data ANalysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.boxplot(x='academic_pressure', y='average_sleep', data=df)
plt.title('Average Sleep by Academic Pressure Level')
plt.xlabel('Academic Pressure')
plt.ylabel('Average Sleep (hours)')
plt.show()
```

This figure illustrates differences in average sleep across levels of academic pressure.

### Relationship between Academic Pressure and Sleep

The boxplot shows that students reporting higher levels of academic pressure tend to have lower average sleep duration.
While there is variability within each group, a general pattern suggests that increased academic pressure is associated
with reduced sleep.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='financial_concerns', y='average_sleep', data=df)
plt.title('Average Sleep by Financial Concerns')
plt.xlabel('Financial Concerns')
plt.ylabel('Average Sleep (hours)')
plt.show()
```

This figure illustrates differences in average sleep across levels of academic pressure.


## Key Findings

- Higher levels of academic pressure are associated with lower average sleep duration among students.
- Financial concerns also appear to be related to reduced sleep, although variability exists across groups.
- These patterns suggest that multiple academic and financial stressors may be linked to sleep behavior in student populations.

## Limitations & Next Steps

- This analysis is based on self-reported survey data, which may be subject to reporting bias.
- The sample size is relatively small and may not be representative of all student populations.
- Future analyses could examine additional factors or use longitudinal data to better understand these relationships.
