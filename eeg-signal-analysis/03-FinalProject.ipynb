## Abstract

Epileptic seizures produce distinctive patterns of abnormal electrical activity in the brain that can be detected through electroencephalography (EEG). Despite multitudes of research, there is no universal analytical standard for differentiating seizure from non-seizure activity, as studies often vary in focus between connectivity-based models and single-channel spectral analyses. This project aims to replicate, quantify, and visualize the core time and frequency domain differences between seizure (ictal) and non-seizure (interictal) EEG signals using interpretable, signal-based methods.

We analyzed an open-source EEG dataset containing labeled seizure and non-seizure segments and computed both statistical (root-mean-square amplitude, variance, spike rate) and spectral features (band power). The data analysis revealed that seizure segments exhibited markedly higher amplitude variability and total signal energy, alongside elevated power in low-to-mid frequency bands (1–30 Hz) and reduced spectral entropy, consistent with excessive neural synchronization during ictal states. These results confirm our hypothesis that seizure EEGs can be significantly distinguished from normal activity using fundamental time and frequency domain features (amplitude variance, dominant frequency bands, etc). Through these results, our project contributes to ongoing efforts toward accessible and universally applicable EEG analytics, with potential impact on real-time seizure detection and the development of accesible neurodiagnostic models.

## Background and Prior Work

Epilepsy is a neurological disorder that affects more than 50 million people worldwide, characterized by sudden bursts of abnormal electrical activity in the brain that cause seizures of varying intensity [<sup>1</sup>](#ref1). Electroencephalography (EEG) plays a crucial role in diagnosing and monitoring epilepsy because it captures the brain’s electrical signals non-invasively and in real time. During seizures, EEG signals show distinctive features—such as sharp spikes, rhythmic discharges, and synchronized bursts—that sharply differ from the stable oscillatory patterns seen during non-seizure periods [<sup>3</sup>](#ref3). Understanding these signal differences is central to advancing seizure detection methods.

Prior research has explored EEG data to determine which features best distinguish seizure activity from normal brain states, using both physiological and computational approaches. For example, Adebimpe et al. (2016) examined resting-state EEG data from children with benign epilepsy with centrotemporal spikes (BECTS) and found altered connectivity in the theta (4–7 Hz), alpha (8–13 Hz), and beta (14–30 Hz) bands compared to healthy controls [<sup>2</sup>](#ref2). Their findings suggest that seizure-prone brains exhibit disrupted functional networks and excessive synchronization across multiple frequency domains. However, because the study focused exclusively on a small pediatric BECTS cohort, its conclusions may not generalize to adult-onset epilepsy or other seizure types. Subsequent reviews have noted that variations in recording methods and electrode placement across studies can introduce confounding factors, limiting reproducibility and external validity [<sup>4</sup>](#ref4) [<sup>5</sup>](#ref5).

Complementary research by Fisher et al. (2014) emphasized that seizure onset and termination are not sharply defined in EEG recordings. Instead, seizures often unfold through gradual transitions characterized by rhythmic patterns in delta (0–3 Hz), theta (4–7 Hz), and alpha (8–12 Hz) bands, interspersed with repetitive sharp spikes [<sup>2</sup>](#ref2). This finding highlights that seizure detection requires analyzing both the amplitude and frequency structure of signals, rather than relying solely on threshold-based classification. However, Fisher’s work was largely qualitative, relying on expert interpretation rather than computational quantification. Later studies expanded upon this by employing wavelet transforms, spectral entropy, and machine learning classifiers to detect seizures automatically [<sup>6</sup>](#ref6) [<sup>7</sup>](#ref7). These approaches provided stronger empirical grounding but introduced new challenges, such as potential overfitting due to limited patient diversity and the lack of consistent validation frameworks [<sup>8</sup>](#ref8).

In conclusion, these studies show that seizure-related EEG signals differ in both amplitude variability and frequency composition, often reflecting excessive neural synchronization during ictal states. Yet, there remains no universal consensus on which analytical approach best captures these differences. Connectivity-based studies focus on interregional coherence, while signal-based models emphasize spectral-temporal fluctuations within single channels. By integrating both perspectives, this project aims to quantitatively and visually characterize EEG differences between seizure and non-seizure states while critically addressing common threats to validity such as dataset homogeneity, overinterpretation of visual patterns, and limited reproducibility in prior work.

## Hypothesis


Compared with non-seizure (interictal/normal) EEG segments, seizure (ictal) segments will exhibit higher time-domain signal energy and variability (specifically higher root-mean-square (RMS) amplitude, greater amplitude variance, and higher spike/peak rate per second) and distinct frequency-domain structure, including higher total power in low-to-mid frequency bands (1–30 Hz), disproportionately elevated theta and beta band power, and lower spectral entropy (more concentrated power around dominant rhythmic components). Together, these differences will allow time and frequency domain features to reliably distinguish seizure from non-seizure segments across subjects and recording conditions.

### Dataset #1: Epileptic Seizure Recognition.csv

The Epileptic Seizure Recognition dataset (originally from the UCI Machine Learning Repository, hosted on Kaggle) contains one-dimensional electroencephalogram (EEG) recordings from 500 individuals, segmented into 11,500 one-second intervals. Each observation consists of 178 sequential voltage readings sampled at 178 Hz, representing the brain’s electrical potential in microvolts (µV) over time. EEG measures the summed electrical activity of cortical neurons, and typical resting amplitudes range from 10–100 µV, while seizure (ictal) activity can exceed several hundred µV with rhythmic spikes and bursts..

Important variables and metrics

- `X1–X178`: EEG voltage samples in microvolts (µV). Each feature corresponds to 1/178 s of the one-second window.

- `y` (target variable): Categorical label identifying the condition under which the EEG segment was recorded.

| Label | Condition | Description |
|----------|----------|----------|
| 1 | Seizure activity | EEG recorded during an epileptic seizure (ictal state) |
| 2 | Tumor region (pathological) | EEG recorded from a brain region containing a tumor |
| 3 | Healthy region (same patient) | EEG recorded from a healthy brain area in a tumor patient |
| 4 | Eyes closed | EEG from a healthy individual with eyes closed |
| 5 | Other | EEG from a healthy individual with eyes open |
    
    All subjects falling in classes 2, 3, 4, and 5 are subjects who did not have epileptic seizure. Only subjects in class 1 have epileptic seizure. 


- Sampling frequency: 178 Hz (i.e., 178 samples per second).

- Duration per segment: 1 s of EEG recording.

From these features, time-domain metrics such as root-mean-square (RMS) amplitude (overall energy), variance (signal dispersion), and spike rate (number of sharp peaks per second) can be computed. In the frequency domain, one can derive power spectral density (PSD) in bands such as delta (1–4 Hz), theta (4–7 Hz), alpha (8–13 Hz), beta (14–30 Hz), and gamma (>30 Hz). In typical EEG, alpha activity dominates relaxed wakefulness, while seizures often show abnormally high power in low-frequency (delta/theta) or mid-frequency (beta) bands and reduced spectral entropy due to rhythmic synchronization.

```python
# Apply low pass filter to the Bonn EEG dataset (40 Hz cutoff)
def low_pass_filter(data: pd.DataFrame, cutoff: float=40, fs: float=173.61, order: int=5, label_col: str='label'):
    # Split features / labels
    X_df = data.drop(columns=[label_col])
    y = data[label_col].to_numpy()

    X = X_df.to_numpy(dtype=float)

    sample_cols = X_df.columns
    sos = butter(order, cutoff, btype='low', fs=fs, output='sos')
    X_filt = sosfiltfilt(sos, X, axis=1)
    df_filt = pd.DataFrame(X_filt, columns=sample_cols, index=data.index)
    df_filt[label_col] = y
    return df_filt

df_bonn_filt = low_pass_filter(df_bonn, cutoff=40, fs=173.61, order=5, label_col='label')

# Plot the first row of the filtered data
x = df_bonn_filt.iloc[0, :-1].values
fs = 173.61  # Sampling frequency
t = np.arange(len(x)) / fs  # Time vector
plt.figure()
plt.plot(t, x)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (µV)")
plt.title(f"EEG Segment 0  (n={x.size} samples): After Low-Pass Filter")
plt.show()

print(df_bonn_filt.shape)
df_bonn_filt.head()
```

### Section 2. Time-Domain Analysis for Both Datasets

Time-domain analysis examines EEG signal characteristics directly in the temporal dimension, without frequency transformation. We compute three key metrics aligned with our hypothesis:

1. **RMS (Root Mean Square) Amplitude**  
   RMS amplitude measures the *average signal power* by emphasizing larger voltage fluctuations. During a seizure, excessive and synchronous neuronal firing produces strong, repetitive bursts of high-amplitude activity. These bursts increase the total electrical energy in the signal over time, causing RMS values to rise significantly in seizure segments. This makes RMS a well-established indicator of heightened neural excitation.

2. **Variance**  
   Variance captures how far the signal fluctuates away from its mean. Under normal conditions, EEG activity stays within moderate ranges and varies irregularly. Seizures trigger hypersynchronous discharges involving thousands of neurons firing rhythmically, which leads to sudden and dramatic voltage deviations. As a result, variance increases sharply during ictal episodes compared to interictal (non-seizure) periods.

3. **Peak-to-Peak Amplitude**  
   Peak-to-peak amplitude is the difference between the highest and lowest voltage within a time window. In healthy states, these peaks and troughs are relatively modest. During a seizure, sharp epileptic spikes and large wave complexes form, driving voltage swings to extreme values. This widens the amplitude range and provides a direct indicator of intense pathological brain activity.


**Hypothesis prediction:** If these features are capturing true seizure-related behavior, then seizure segments should show **higher RMS, higher variance, and larger peak-to-peak amplitude** across both datasets.


In both datasets, these three time-domain features give a simple but informative first look at how seizure and non-seizure segments differ. By summarizing each 1-second segment with RMS amplitude, variance, and peak-to-peak amplitude, we compress the raw waveform into interpretable statistics that can be directly compared across classes. This allows us to quickly see whether seizure segments tend to occupy a different range of values than non-seizure segments, even before applying any machine learning models.

#### Section 2a. Epileptic Dataset Time-Domain Analysis

```python
# Compute RMS amplitude and variance for each segment in Epileptic Seizure Recognition Dataset
df_epileptic['rms_amplitude'] = np.sqrt(np.mean(df_epileptic[voltage_cols].values**2, axis=1))
df_epileptic['variance'] = np.var(df_epileptic[voltage_cols].values, axis=1)
df_epileptic['peak_to_peak'] = df_epileptic[voltage_cols].max(axis=1) - df_epileptic[voltage_cols].min(axis=1)

# Visualize distributions
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.boxplot(data=df_epileptic, x='Seizure', y='rms_amplitude', ax=axes[0])
axes[0].set_title('RMS Amplitude by Class')
axes[0].set_ylabel('RMS Amplitude (µV)')
axes[0].set_xlabel('Non-Seizure (0) vs Seizure (1)')

sns.boxplot(data=df_epileptic, x='Seizure', y='variance', ax=axes[1])
axes[1].set_title('Variance by Class')
axes[1].set_ylabel('Variance (µV²)')
axes[1].set_xlabel('Non-Seizure (0) vs Seizure (1)')

sns.boxplot(data=df_epileptic, x='Seizure', y='peak_to_peak', ax=axes[2])
axes[2].set_title('Peak-to-Peak Amplitude by Class')
axes[2].set_ylabel('Peak-to-Peak (µV)')
axes[2].set_xlabel('Non-Seizure (0) vs Seizure (1)')

plt.tight_layout()
plt.show()
```

- Minimum spacing: 20ms between peaks (prevents double-counting)

**Interpretation:** Spike rate quantifies the frequency of sharp transient events, which are hallmarks of epileptiform activity. Higher rates indicate more frequent pathological discharges.

### Section 4: Spectral Entropy Analysis

**Spectral entropy** measures the disorder or randomness in the frequency-domain representation of the EEG signal. It is computed as the Shannon entropy of the normalized power spectrum:

**Formula:** H = -Σ(p_i × log₂(p_i))

where p_i is the normalized power at frequency i.

**Interpretation:**
- **High entropy**: Power is distributed evenly across many frequencies (random, irregular activity)
- **Low entropy**: Power is concentrated in few frequencies (synchronized, rhythmic activity)

**Hypothesis prediction:** Seizures will show **lower spectral entropy** because pathological neural synchronization concentrates power in specific frequency bands, creating more ordered, rhythmic patterns rather than the broad, irregular spectrum of normal brain activity.

```python
# For Epileptic Seizure Recognition Dataset, compute spectral entropy
spectral_entropy_data = []
for idx, row in df_epileptic.iterrows():
    signal_data = row[voltage_cols].values
    spec_entropy = compute_spectral_entropy(signal_data, fs=178)
    spectral_entropy_data.append(spec_entropy)

df_epileptic['spectral_entropy'] = spectral_entropy_data
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# Boxplot
sns.boxplot(data=df_epileptic, x='Seizure', y='spectral_entropy', ax=axes[0])
axes[0].set_title('Spectral Entropy by Class (Epileptic Dataset)', fontsize=14)
axes[0].set_xlabel('Non-Seizure (0) vs Seizure (1)')
axes[0].set_ylabel('Spectral Entropy')
# Histogram
seizure_entropy_epileptic = df_epileptic[df_epileptic['Seizure'] == 1]['spectral_entropy']
nonseizure_entropy_epileptic = df_epileptic[df_epileptic['Seizure'] == 0]['spectral_entropy']
axes[1].hist(nonseizure_entropy_epileptic, bins=30, alpha=0.6, label='Non-Seizure', color='blue', density=True)
axes[1].hist(seizure_entropy_epileptic, bins=30, alpha=0.6, label='Seizure', color='red', density=True)
axes[1].set_xlabel('Spectral Entropy')
axes[1].set_ylabel('Density')
axes[1].set_title('Spectral Entropy Distribution (Epileptic Dataset)', fontsize=14)
axes[1].legend()
```

### Section 5: Frequency-Domain Analysis

Frequency-domain analysis decomposes EEG signals into their constituent frequency components using Fast Fourier Transform (FFT). We examine power in five clinically-defined EEG bands:

| Band | Frequency Range | Normal Association |
|------|----------------|-------------------|
| Delta | 1-4 Hz | Deep sleep, pathological conditions |
| Theta | 4-7 Hz | Drowsiness, meditation, some pathologies |
| Alpha | 8-13 Hz | Relaxed wakefulness, eyes closed |
| Beta | 14-30 Hz | Active thinking, concentration |
| Gamma | 30-89 Hz | High-level cognitive processing |

**Hypothesis prediction:** Seizures will show elevated power in low-to-mid frequency bands (delta, theta, beta) and disproportionately elevated theta and beta relative to alpha, reflecting the characteristic rhythmic discharges of ictal activity.

**Why log scale:** Band powers span several orders of magnitude, so we use logarithmic y-axes for clearer visualization of all bands simultaneously.

Applying this frequency-domain framework to our datasets, we can observe how the spectral content of seizure segments differs from that of non-seizure segments. When we compare the distributions of log-band power, seizure segments tend to show a relative increase in lower-frequency bands (especially delta and theta) and a decrease in alpha power, reflecting a shift toward slower, more synchronized rhythms during ictal events. Differences in beta and gamma power are also visible, although the separation between classes is generally less pronounced than in the lower bands.

### Section 6: Band Power Ratios Analysis

While absolute band powers show significant differences between seizure and non-seizure states, **band power ratios** provide a more robust measure of frequency distribution that accounts for inter-subject variability and overall signal amplitude differences. Our hypothesis specifically mentions "**disproportionately elevated theta and beta band power**" - ratios directly test this claim by normalizing band powers relative to each other.

**Key Ratios:**
- **Theta/Alpha Ratio**: Indicates shift toward slower theta rhythms relative to alpha
- **Beta/Alpha Ratio**: Indicates shift toward faster beta activity relative to alpha
- **Low-Freq/High-Freq Ratio**: Overall balance between slow (delta+theta) and fast (alpha+beta) rhythms
- **Relative Powers**: Proportion of total power in each band (normalized measure)

These ratios are clinically meaningful and commonly used in EEG analysis to characterize brain states independent of overall signal amplitude.

```python
from eeg_features import compute_band_ratios

# Compute band power ratios for Epileptic Dataset
ratio_data_epileptic = []
for idx, row in df_epileptic.iterrows():
    signal_data = row[voltage_cols].values
    band_powers, _, _ = compute_band_power(signal_data, fs=178)
    ratios = compute_band_ratios(band_powers)
    ratios['class'] = row['Seizure']
    ratio_data_epileptic.append(ratios)

df_ratios_epileptic = pd.DataFrame(ratio_data_epileptic)

# Visualize key ratios
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()
fig.suptitle('EEG Band Power Ratios by Class (Epileptic Dataset)', fontsize=16)

key_ratios = ['theta_alpha', 'beta_alpha', 'low_high', 'theta_relative', 'beta_relative', 'alpha_relative']
ratio_labels = ['Theta/Alpha Ratio', 'Beta/Alpha Ratio', 'Low-Freq/High-Freq Ratio', 
                'Theta Relative Power', 'Beta Relative Power', 'Alpha Relative Power']

for idx, (ratio, label) in enumerate(zip(key_ratios, ratio_labels)):
    sns.boxplot(data=df_ratios_epileptic, x='class', y=ratio, ax=axes[idx])
    axes[idx].set_title(f'{label} (Epileptic Dataset)', fontsize=12)
    axes[idx].set_ylabel(label)
    axes[idx].set_xlabel('Non-Seizure (0) vs Seizure (1)')

plt.tight_layout()
plt.show()
```

```python
print("=" * 80)
print("BAND POWER RATIO ANALYSIS - EPILEPTIC DATASET")
print("=" * 80)
print("\nTesting hypothesis: Seizures show 'disproportionately elevated' theta and beta power")
print("=" * 80)

for ratio, label in zip(key_ratios, ratio_labels):
    seizure_ratio = df_ratios_epileptic[df_ratios_epileptic['class'] == 1][ratio]
    nonseizure_ratio = df_ratios_epileptic[df_ratios_epileptic['class'] == 0][ratio]
    
    # Mann-Whitney U test
    stat, p_value = mannwhitneyu(seizure_ratio, nonseizure_ratio, alternative='two-sided')
    
    # Cohen's d
    d = cohens_d(seizure_ratio, nonseizure_ratio)
    
    # Compute statistics
    seizure_mean = seizure_ratio.mean()
    nonseizure_mean = nonseizure_ratio.mean()
    percent_diff = ((seizure_mean - nonseizure_mean) / (nonseizure_mean + 1e-10)) * 100
    
    print(f"\n{label}:")
    print(f"  Non-Seizure: {nonseizure_mean:.4f} ± {nonseizure_ratio.std():.4f}")
    print(f"  Seizure: {seizure_mean:.4f} ± {seizure_ratio.std():.4f}")
    print(f"  Difference: {percent_diff:+.1f}%")
    print(f"  Mann-Whitney U: {stat:.4f}, p-value: {p_value:.4e}")
    print(f"  Cohen's d: {d:.4f}")
    print(f"  Significance: {'✓ Yes (p < 0.05)' if p_value < 0.05 else '✗ No'}")
    print(f"  Effect size: {('Large' if abs(d) > 0.8 else 'Medium' if abs(d) > 0.5 else 'Small' if abs(d) > 0.2 else 'Negligible')}")
    
    # Interpretation based on ratio type
    if 'theta_alpha' in ratio:
        print(f"  → {'CONFIRMS' if seizure_mean > nonseizure_mean and p_value < 0.05 else 'Does not confirm'} disproportionate theta elevation")
    elif 'beta_alpha' in ratio:
        print(f"  → {'CONFIRMS' if seizure_mean > nonseizure_mean and p_value < 0.05 else 'Does not confirm'} disproportionate beta elevation")
    elif 'low_high' in ratio:
        print(f"  → {'Shows' if seizure_mean > nonseizure_mean else 'No'} shift toward low-frequency dominance")

print("\n" + "=" * 80)
print("SUMMARY: Hypothesis Validation")
print("=" * 80)
print("The band power ratio analysis directly tests the 'disproportionately elevated'")
print("claim in our hypothesis. Ratios normalize for overall amplitude differences.")
```

- **Theta/Alpha and Beta/Alpha ratios are HIGHER** in seizures (p < 0.001), **confirming** the "disproportionately elevated" claim

- **Low/High ratio is LOWER** in seizures (-43%), contradicting simple "low-frequency dominance" idea

### Key Insights

**1. Broadband vs. Selective Activation**
- Initial hypothesis: seizures concentrate power in specific bands (lower entropy)
- **Actual finding**: seizures elevate power across ALL bands simultaneously (higher entropy)
- **Implication**: Seizures represent global cortical hypersynchronization, not selective frequency enhancement

**2. Amplitude-Independent Discrimination**
- Absolute features (RMS, variance) show largest effects (d > 2.0)
- Relative features (ratios) show smaller but significant effects (d = 0.2-1.0)
- **Implication**: Both absolute and relative features are valuable; ratios are more robust to recording variability

## Discusison and Conclusion

Our project built on multiple prior research works showing that seizure-related EEG activity differs markedly from normal brain activity in both amplitude and frequency composition. Since past studies diverged on which analytical framework(specifically connectivity-based or signal-based) most effectively captures these distinctions, our approach focused on quantifying time and frequency-domain patterns from single-channel signals to find reproducible seizure signatures. We combined amplitude-variance measures with frequency-based power analysis, bridging the gap between purely spectral and purely temporal perspectives discussed in our background section. Although our analysis did not introduce new modeling frameworks or connectivity metrics, it validated that straightforward statistical and spectral descriptors remain highly effective and interpretable for seizure characterization.  

Our exploratory data analysis provided clear evidence that seizure (ictal) EEG segments differ significantly from non-seizure (interictal) segments in their time-domain characteristics. Specifically, seizure samples showed substantially higher root-mean-square (RMS) amplitudes, greater variance, and a higher rate of transient spikes. These features collectively indicate elevated signal energy and instability, consistent with the excessive neural synchronization commonly associated with epileptic discharges. These findings support the time-domain part of our hypothesis, confirming that signal amplitude and variability are reliable indicators of seizure activity. For frequency, our analysis also revealed that seizure segments exhibited higher overall power, particularly in the low-to-mid frequency ranges (1–30 Hz), encompassing the delta, theta, and beta bands. The power in these bands was not only elevated but also more concentrated, suggesting reduced spectral entropy during seizure episodes. We ran several tests and calculated statistical values on both datasets such as chi-squared, Mann-Whitney, and Cohen's d while plotting dominant frequency by class, which all confirmed strong association and significant difference between seizure and non-seizure classes This observation aligns with past findings that seizures are characterized by synchronized slow-wave repetitions and reduced spectral complexity, confirming the frequency-domain part of our hypothesis. 

However, not all results were cleanly separable or observable. Some non-seizure segments displayed short bursts of heightened amplitude or low-frequency activity resembling ictal patterns, while a few seizure-labeled segments appeared overly stable. These inconsistencies likely stem from labeling variability, physiological differences between patients, or personal/circumstantial differences that cause confusion. Consequently, while our results confirm the hypothesized differences in general, they also reveal that seizure and non-seizure EEGs exist on a spectrum rather than as entirely distinct clusters that are clear to observe. This means there is still a need for patient-specific thresholds when developing automated seizure detection systems.

There were several limitations to our research. The first came with our datasets, as our study’s scope was constrained by dataset homogeneity and lack of longitudinal data. The EEG recordings came from a limited subject pool (one dataset had only 10 subjects), which restricts generalizability across demographics and seizure subtypes. Furthermore, the dataset’s cross-sectional nature prevented us from examining dynamic transitions between pre/post ictal and ictal states (before, during, and after seizure activities), an important factor in understanding seizure onset and recovery. Future work could expand upon this foundation by having or using a larger and more diverse dataset, integrating multi-channel spatial coherence analyses, and having or using models that capture nonlinear interactions as well.

In conclusion, our research and analysis confirm the central aspects of our hypothesis: seizure EEGs exhibit higher amplitude, greater variability, and elevated power in low-to-mid frequency bands compared with non-seizure segments. The results carry meaningful implications for the neurology and neurotechnology industries and can be extended further than simple classroom settings. The ability to characterize seizure activity through interpretable time and frequency domain EEG features can be the foundation for innovative products such as lightweight or even wearable models for early seizure detection in hospitals, homes, and even our bodies. These models and devices could allow real-time identification of abnormal neural activity without having to go through large-scale, time-consuming tests, providing efficient and accessible health technology to those who need them.
